# Price Preprocess

```python
import numpy as np
import pandas as pd
from pykalman import KalmanFilter


def kalman_trend(price):

    y = np.log(price).values

    kf = KalmanFilter(

        # state = [level, slope]

        transition_matrices=np.array([
            [1, 1],
            [0, 1]
        ]),

        observation_matrices=np.array([
            [1, 0]
        ]),

        transition_covariance=np.array([
            [1e-5, 0],
            [0, 1e-6]
        ]),

        observation_covariance=1e-3,

        initial_state_mean=np.array([
            y[0],
            0
        ]),

        initial_state_covariance=np.eye(2)
    )

    state_means, state_covs = kf.filter(y)

    df = pd.DataFrame(
        index=price.index,
        data={
            "log_price": y,
            "trend": state_means[:, 0],
            "slope": state_means[:, 1]
        }
    )

    return df
```

```python
def _get_kf_px(df: pd.DataFrame) -> pd.DataFrame:

    df_out = kalman_trend(df.px_val)
    return df_out

path  = r"C:\Users\Diego\Documents\GitHub\TrendSystem\data\FuturesData\PrepFuturesPX.parquet"
df_kf = (pd.read_parquet(
    path = path, engine = "pyarrow").
    set_index("date").
    groupby("ticker").
    progress_apply(lambda group: _get_kf_px(group)))
```

```python
def _tmp_calc(df: pd.DataFrame, phi: float = 0.98) -> pd.DataFrame: 
    
    df_out = (df.drop(
        columns = ["px_diff", "px_dt", "signal", "lag_signal"]).
        assign(
            pred       = lambda x: x.trend + (phi * x.slope),
            signal     = lambda x: (x.pred - x.log_price),
            lag_signal = lambda x: x.signal.shift(),
            rtn        = lambda x: x.price.pct_change(),
            signal_rtn = lambda x: -np.sign(x.lag_signal) * x.rtn))

    return df_out

model4_path = r"C:\Users\Diego\Documents\GitHub\TrendSystem\data\KalmanFilter\Model4LogPrices.parquet"
(pd.read_parquet(
    path = model4_path, engine = "pyarrow").
    groupby("ticker").
    apply(_tmp_calc).
    reset_index().
    pivot(index = "date", columns = "ticker", values = "signal_rtn").
    apply(lambda x: x * (0.1 / (x.ewm(span = 100, adjust = False).std() * np.sqrt(252)))).
    apply(lambda x: np.where(np.abs(x) > 2, np.nan, x)).
    mean(axis = 1).
    to_frame(name = "tmp").
    apply(lambda x: np.cumprod(1 + x)).
    plot(logy = True))
```

```python
df_signal_rtn = (df_rtn.merge(
    right = df_signal, how = "inner", on = ["date", "ticker"]).
    assign(signal_rtn = lambda x: np.sign(x.lag_signal) * x.rtn))
```

```python
df_cuml = (df_signal_rtn.pivot(
    index = "date", columns = "ticker", values = "signal_rtn").
    apply(lambda x: x * (0.2 / (x.ewm(span = 100, adjust = False).std() * np.sqrt(252)))).
    rolling(window = 5).sum().
    reset_index().
    melt(id_vars = "date", value_name = "cuml_rtn").
    dropna())
```

```python
df_combined = (df_signal_rtn[
    ["date", "ticker", "lag_signal"]].
    merge(right = df_cuml, how = "inner", on = ["date", "ticker"]))
```
