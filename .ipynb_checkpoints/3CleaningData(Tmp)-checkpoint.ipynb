```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Cleaning Data

```python
data_path   = os.path.join(os.getcwd(), "data")
fut_path    = os.path.join(data_path, "RawFuturesPX.parquet")
spot_path   = os.path.join(data_path, "EquityIndexSpotPrices.xlsx")
ticker_path = os.path.join(data_path, "TickerGuide.xlsx") 
```

```python
df_spot_px = (pd.read_excel(
    io = spot_path).
    rename(columns = {"Date": "date"}).
    melt(id_vars = "date", var_name = "ticker", value_name = "spot_px"))
```

```python
df_fut_px = (pd.read_parquet(
    path = fut_path, engine = "pyarrow"))
```

```python
tickers = {"SMI": "SM1 Index"}
for ticker in tickers.keys():

    df_spot_tmp = (df_spot_px.query(
        "ticker == @ticker").
        set_index("date"))

    fut_ticker = tickers[ticker]
    df_fut_tmp = df_fut_px.query("ticker == @fut_ticker")
    break
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path     = r"C:\Users\Diego\Desktop\WeekyNotebooks\20260705TmpFutures\EquityFutures"
files    = [file for file in os.listdir(path) if file[0:3] == "Fut"]
df_lists = []

for file in files: 

    tmp_path = os.path.join(path, file)
    df_raw   = (pd.read_excel(
        io         = tmp_path, 
        sheet_name = "PX_LAST").
        rename(columns = {"Date": "date"}).
        set_index("date"))

    df_raw.columns = [col.split("(")[0].strip() for col in df_raw.columns]
    
    df_tmp = (df_raw.reset_index().melt(
        id_vars = "date", var_name = "ticker", value_name = "PX_LAST").
        dropna())

    df_lists.append(df_tmp)

df_out = pd.concat(df_lists)
```

```python
tickers = df_out.ticker.drop_duplicates().sort_values().to_list()
```

```python
path = r"C:\Users\Diego\Documents\GitHub\TrendSystem\data\FuturesData\RawFuturesPX.parquet"
df_tmp =(pd.read_parquet(
    path = path, engine = "pyarrow").
    query("ticker == @tickers"))
```

```python
tickers_ = df_out.ticker.drop_duplicates().sort_values().to_list()
tickers  = [ticker.replace("COMB","").strip() for ticker in tickers_]
```

```python
for ticker in tickers: 

    display(df_tmp.query(
        "ticker == @ticker").
        set_index("date")
        [["PX_LAST"]].
        sort_index().
        plot(title = ticker))

    plt.show()
```

```python
def _get_fx(df: pd.DataFrame) -> pd.DataFrame: return df.query("date == date.max()").drop_duplicates()

path = r"C:\Users\Diego\Documents\GitHub\TrendSystem\data\FuturesData\PrepFuturesPX.parquet"
(pd.read_parquet(
    path = path, engine = "pyarrow")
    [["date", "fx_ticker", "fx_val"]].
    groupby("fx_ticker").
    apply(_get_fx).
    reset_index()
    [["fx_ticker", "fx_val"]])
```

```python
(pd.read_parquet(
    path = path, engine = "pyarrow").
    query("fx_val != fx_val").
    fx_ticker.
    drop_duplicates().
    sort_values().
    to_list())
```
