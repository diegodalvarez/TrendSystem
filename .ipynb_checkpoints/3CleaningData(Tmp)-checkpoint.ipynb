# Price Preprocess

```python
import numpy as np
import pandas as pd
from pykalman import KalmanFilter


def kalman_trend(price):

    y = np.log(price).values

    kf = KalmanFilter(

        # state = [level, slope]

        transition_matrices=np.array([
            [1, 1],
            [0, 1]
        ]),

        observation_matrices=np.array([
            [1, 0]
        ]),

        transition_covariance=np.array([
            [1e-5, 0],
            [0, 1e-6]
        ]),

        observation_covariance=1e-3,

        initial_state_mean=np.array([
            y[0],
            0
        ]),

        initial_state_covariance=np.eye(2)
    )

    state_means, state_covs = kf.filter(y)

    df = pd.DataFrame(
        index=price.index,
        data={
            "log_price": y,
            "trend": state_means[:, 0],
            "slope": state_means[:, 1]
        }
    )

    return df
```

```python
def _get_kf_px(df: pd.DataFrame) -> pd.DataFrame:

    df_out = kalman_trend(df.px_val)
    return df_out

path  = r"C:\Users\Diego\Documents\GitHub\TrendSystem\data\FuturesData\PrepFuturesPX.parquet"
df_kf = (pd.read_parquet(
    path = path, engine = "pyarrow").
    set_index("date").
    groupby("ticker").
    progress_apply(lambda group: _get_kf_px(group)))
```

```python
df_signal = (df_kf.reset_index().pivot(
    index = "date", columns = "ticker", values = "trend").
    diff().
    apply(lambda x: x / x.ewm(span = 10, adjust = False).std()).
    apply(lambda x: x.ewm(span = 100, adjust = False).mean()).
    reset_index().
    melt(id_vars = "date", value_name = "signal").
    dropna())
```

```python
path = r"C:\Users\Diego\Documents\GitHub\TrendSystem\data\FuturesData\PrepFuturesPX.parquet"
df_rtn = (pd.read_parquet(
    path = path, engine = "pyarrow").
    pivot(index = "date", columns = "ticker", values = "px_val").
    pct_change().
    reset_index().
    melt(id_vars = "date", value_name = "rtn").
    dropna())
```

```python
def _tmp_calc(df: pd.DataFrame) -> pd.DataFrame: 
    
    df_out = (df.drop(
        columns = ["px_diff", "px_dt", "signal", "lag_signal"]).
        assign(
            pred       = lambda x: x.trend + x.slope,
            signal     = lambda x: (x.pred - x.log_price),
            lag_signal = lambda x: x.signal.shift(),
            rtn        = lambda x: x.price.pct_change(),
            signal_rtn = lambda x: np.sign(x.lag_signal) * x.rtn)

model4_path = r"C:\Users\Diego\Documents\GitHub\TrendSystem\data\KalmanFilter\Model4LogPrices.parquet"
(pd.read_parquet(
    path = model4_path, engine = "pyarrow").
    groupby("ticker").
    apply(_tmp_calc))
```
